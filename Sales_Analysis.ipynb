```python
state_sales=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
state_sales_amt=df.groupby(['State','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
state_sales_amt
sns.set(rc={'figure.figsize':(20,8)})# first argument is for width and second is for height
# above line is mainly for the gap between the bars and the x axis values so it will not get overlaped
ax=sns.barplot(data=state_sales_amt,x='State',y='Amount',hue='Gender')
```

# Marital Status


```python
ax=sns.countplot(data=df,x='Marital_Status')
#countplot is used to count of values of x axis
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

from the above graph we can see that the most of the buyers count are married (women) 

```python
state_sales_amt=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
ax=sns.barplot(data=state_sales_amt,x='Marital_Status',y='Amount',hue='Gender')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

from the above graph we can see that the most of the buyers are married (women) and they have very high purchasing power

# Product_category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

from the above graph we can see that the most of the sold product are from the food , clothing and appareal and electronics
in top 10 product category

```python
df.columns
```

```python
top_selling_prd=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(data=top_selling_prd,x='Product_ID',y='Orders')

for bars in ax.containers:
    ax.bar_label(bars)
 
```
